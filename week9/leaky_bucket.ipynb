```python
import time

bsize = 512

def bucketInput(a,b):
    if a > bsize:
        print("Bucket overflow")
    else:
        time.sleep(1)
        while a > b:
            print(b," bytes outputed")
            a = a-b
            time.sleep(1)
        if a > 0:
            print("Last ",a," bytes sent")
        print("Bucket output successful")
```

```python
import random

def main():
    print("Enter the number of packets")
    n = int(input())
    print("Enter the output rate")
    op = int(input())
    for i in range(n):
        time.sleep(random.randrange(0,10))
        psize = random.randrange(0,1000)
        print("Packet no: ",i," Packet size: ",psize)
        bucketInput(psize,op)
```
